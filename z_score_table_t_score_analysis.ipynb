# Create a z-score table


```python
import numpy as np
import pandas as pd
from scipy.integrate import quad
```

```python
def normal_probability_density(x):
    constant = 1.0 / np.sqrt(2 * np.pi)
    return(constant * np.exp((-x**2) / 2))
```

```python
standard_normal_table = pd.DataFrame(data=[],
                                     index = np.round(np.arange(0, 3.5, .1), 2),
                                     columns=np.round(np.arange(0.00, 0.1, 0.01), 2))

for index in standard_normal_table.index:
  for column in standard_normal_table.columns:
    z = np.round(index + column, 2)
    value, _ = quad(normal_probability_density, np.NINF, z)
    standard_normal_table.loc[index, column] = value

standard_normal_table.index = standard_normal_table.index.astype(str)
standard_normal_table.columns = (str(column).ljust(4, '0') for column in standard_normal_table.columns)

standard_normal_table
```

# T-student test from two datasets

```python
# packages

from math import sqrt
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from scipy.stats import sem
from scipy.stats import t
```

```python
# input data

seed(1)
dataset_1 = 6 * randn(300) + 30
dataset_2 = 6 * randn(300) + 40
alpha = 0.05
```

```python
# t_test function calculator

def t_test(dataset_1: list, dataset_2: list, alpha: float) -> float:
  """Calculate T test from two dataset
  dataset_1: [list] the first dataset
  dataset_2: [list] the second dataset

  return: [int] the p-value for the distributions
  """
  mean_1, mean_2 = mean(dataset_1), mean(dataset_2)

  stand_err_1, stand_err_2 = sem(dataset_1), sem(dataset_2)

  stand_err_diff = sqrt(stand_err_1**2 + stand_err_2**2)

  t_stat = (mean_1 - mean_2) / stand_err_diff

  df = (len(dataset_1) + len(dataset_2)) - 2

  critical = t.ppf(1 - alpha, df)

  p_value = (1 - t.cdf(abs(t_stat), df)) * 2

  return t_stat, df, critical, p_value
```

```python
# solve for input values

t_stat, df, critical, p_value = t_test(dataset_1, dataset_2, alpha)

print(f"t_stat = {t_stat:.2f}, df = {df}, critical = {critical:.2f}, p-value = {p_value:.4f}")
```

```python
# critical value test

if abs(t_stat) <= critical:
  print("accept H0")
else:
  print("reject H0")
```

```python
# p-value test
if p_value > alpha:
    print("accept H0")
else:
  print("reject H0")
```
